```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/zomato-bangalore-restaurants/zomato.csv')
```

```python
df.columns
```

```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

Dropping Duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

Cleaning of Data

```python
df['rate'].unique()
```

```python
#loading only those data which are having numbers after removing the "/5" and any other text
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

Filling Null values in rate column with Mean

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping Null Values

```python
df.dropna(inplace=True)
df.head()
```

```python
#Renaming Columns 
df.rename(columns={'approx_cost(for two people)':'Cost for Two','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

Since the names of listed_in(city) are already available in locations so we can drop the column listed_in(city)

```python
df=df.drop(['listed_in(city)'],axis=1)
df.head()
```

```python
df['Cost for Two'].unique()
```

```python
#removing comma from column Cost for Two
def removecomma(value):
    value=str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost for Two']=df['Cost for Two'].apply(removecomma)
df['Cost for Two'].unique()
```

```python
df.head()
```

```python
#Cleaning rest_type column
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

Changing Rest_Types less than 1000 in frequency as Others

```python
def manage_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(manage_rest_type)
df['rest_type'].value_counts()
```

```python
#Cleaning Location column
df['location'].value_counts()
```

```python
#Changing all Locations less than 300 times as others
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]
```

```python
def manage_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location']=df['location'].apply(manage_location)
df['location'].value_counts()
```

```python
df.head()
```

#Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending=False)
cuisines_lessthan100 = cuisines[cuisines<100]
```

```python
def manage_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(manage_cuisines)
df['cuisines'].value_counts()
```
